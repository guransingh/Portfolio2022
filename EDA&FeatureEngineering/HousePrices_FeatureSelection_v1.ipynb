```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

# Step option to show max columns
pd.pandas.set_option('display.max_columns',None)
```

```python
#import dataset
dataset=pd.read_csv('/Users/guransingh/Library/CloudStorage/OneDrive-Personal/Documents/Work/Python Notebooks/Python Notebooks/HousePrices/x_train.csv')
```

### Step 1 - Split into x and y variables

```python
#select y variable
y_train=dataset.iloc[:,-1]
```

```python
#check y head
y_train.head(2)
```

```python
#select x variable
x_train=dataset.iloc[:,:-1]
```

```python
#check xhead
x_train.head(2)
```

### Step 2 - Apply feature selection

```python
"""
Apply feature selection.
First, select the alpha value (the amount of lasso penalty - see your onenote doc)
Larger the alpha, fewer the features that will be selected

Then use SelectfromModel from SKlearn which will select the features which have coefficients non-zero
"""

feature_sel_model = SelectFromModel(Lasso(alpha=0.005, random_state=0))
```

### Step 3 - Fit the model

```python
#fit the model on x train and y train
feature_sel_model.fit(x_train,y_train)
```

```python
#this shows which features are selected and which are not. If alpha is bigger, fewer will be selected
feature_sel_model.get_support()
```

```python
#now check which features were selected
selected_features = x_train.columns[(feature_sel_model.get_support())]
```

```python
#now print the number of selected features
print('Selected features are:',len(selected_features))
print('Total features are:',len(x_train.columns))
selected_features
```

```python
#check the original x_train shape
x_train.shape
```

```python
#now transform the x train
x_train=feature_sel_model.transform(x_train)
```
